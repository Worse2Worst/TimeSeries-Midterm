```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET
from xml.dom import minidom
from itertools import repeat
from IPython.display import Image
import re
```

# Read XML files

```python
# path = 'musicXML/'
# path += 'TheTheme'
# path+= 'BilliesBounce'
# path+='BagsGroove'
# path+='Dexterity'
# path += 'Anthropology'
# path+='C-JamBlues'
path ='All Blues'
# path+='Oleo'
# path +='FlyMe'
# path +='SomedayMyPrinceWillCome'
# path += 'StraightNoChaser'
# path += 'ATrain'
# path += 'SoWhat'
# path += 'Ipanema'

tree = ET.parse(path)
root = tree.getroot()


# Store bars(measures) in bars
bars =[]
chords = []

for elem in root:  
    if(elem.tag=='part'):
        chords = elem
len(chords)
```

# Get the structure,coda and put it in 'bars'

```python
i = -1
start = -999
repeat = -999
end1 = -999
end2 = -999

for bar in chords:
    i+=1
    for elem in bar:
        if(elem.tag=='barline'):
            for marker in (elem):
                if(marker.tag=='repeat'):
                    if(marker.attrib['direction']=='forward'):
                        start = i
                    elif(marker.attrib['direction']=='backward'):
                        repeat = i
```

# continue from the last cell

```python
i = -1
for bar in chords:
    i+=1
    for elem in bar:
        if(elem.tag=='barline'):
            for marker in (elem):
                if(marker.tag=='ending'):
                    if(marker.attrib['type']=='discontinue'):
                        if(marker.attrib['number']=='1'):
                            end1 = i
                        elif(marker.attrib['number']=='2'):
                            end2 = i
```

# Join them together, make the bars of the whole song 

```python
if(start>=0):
    bars = chords[start:(repeat+1)]+chords[start:end1] + chords[end2:]
else:
    bars = chords
```

# Extract the chords, 'Half-Dim's have to be processed carefully

```python
i = 0
# for bar in bars:
#     i+=1
#     for elem in bar:
#         if(elem.tag=='barline'):
#             for marker in (elem):
#                 if(marker.tag=='repeat'):
#                     if(marker.attrib['direction']=='forward'):
#                         start = i
#                     elif(marker.attrib['direction']=='backward'):
#                         repeat = i


# for elem in bars[0]:
#     print(elem.tag)
#     for subelem in elem:
#         print(subelem.tag)
chords = []
durations = []
i = 0
for bar in bars:
    
    for harmony in bar.findall('harmony'):
        chordSymbol = harmony[0][0].text
        accidental =  harmony[0][1].text
        chordType = harmony[1].text
        if(len(harmony)>2 and harmony[2].tag=='degree'):
            deg = harmony[2]
            val = deg[0].text
            alt = deg[1].text
            if(val =='5' and alt =='-1'):
                chordType = 'half-dim'
        chords.append([chordSymbol,accidental,chordType])
    for note in bar.findall('note'):
        for dur in note.findall('duration'):
            durations.append(dur.text)
#             print(dur.text)
    i+=1

durations = [int(dur)/768 for dur in durations]
```


# The Final BARS of the whole song

```python
BARS = [[x[0]]*int(x[1]) for x in list(zip(chords,durations)) ]
```

# Map from (chords -> numbers), using Cycle of Fifths

![title](images/cir5.png)

```python
flat = {'C': -7, 'D': -5,'E': -3,'F': -8,'G': -6,'A': -4,'B': -2}
natural = {'C': 0, 'G':1,'D':2,'A':3,'E':4,'B':5,'F':-1}
sharp = {'C': 7, 'G':8,'D': 9,'E': 11,'F': 6,'A': 11,'B': 12}

```

# Map from ChordTypes -> Intergers

Multiplier = 30  
Major -> 0  
Minor -> -1  
Dominant -> +1  
Half-Diminished -> -2  
Augmented -> +2  
Diminished -> -3  

```python
multiplier = 30
```

```python
graph = []
for bar in BARS:
    for [chord,accidental,chordType] in bar:
        if(accidental == '-1'):
            num = flat[chord]
            print(str(chord)+':'+str(num))
        elif(accidental == '1'):
            num = sharp[chord]
            print(str(chord)+':'+str(num))
        else:
            num = natural[chord]
            print(str(chord)+':'+str(num))
```

# Get the GRAPH!!!

```python
multiplier = 30
for bar in BARS:
    for [chord,accidental,chordType] in bar:
        if(accidental == '-1'):
            num = flat[chord]
            print(str(chord)+':'+str(num))
        elif(accidental == '1'):
            num = sharp[chord]
            print(str(chord)+':'+str(num))
        else:
            num = natural[chord]
        if('major' in chordType):
            num += (0 * multiplier)
        elif('minor' in chordType):
            num += (-1 * multiplier)
        elif('dominant' in chordType):
            num += (1 * multiplier)
        elif('half-dim' in chordType):
            num += (-2 * multiplier)
        elif('diminished' in chordType):
            num += -3 * multiplier
        elif('augment' in chordType):
            num += 2 * multiplier
        graph.append(num)
```
